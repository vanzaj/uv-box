## FizzBuzz

````python
def fizzbuzz(num: int, config: Dict[int, str] | None = None) -> str:
    if config is None:
        config = {3: "Fizz", 5: "Buzz", 15: "FizzBuzz"}
    config = sorted(config.items(), key=lambda x: x[0], reverse=True)
    for key, value in config:
        if num % key == 0:
            return value

    return str(num)
````

````python
%%ipytest -vv


@pytest.mark.parametrize("num,output", [
    (1, "1"),
    (2, "2"),
    (3, "Fizz"),
    (6, "Fizz"),
    (5, "Buzz"),
    (10, "Buzz"),
    (15, "FizzBuzz"),
])
def test_fizzbuzz(num, output):
    assert fizzbuzz(num) == output


def test_fizzbuzz_with_config():
    config = {7: "Bugs"}
    assert fizzbuzz(7, config) == "Bugs"
````

## TDD for an algo

Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target. You may assume that each input would have exactly one solution, and you may not use the same element twice. You can return the answer in any order.

```
Example 1:
Input: nums = [2,7,11,15], target = 9
Output: [0,1]
Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].

Example 2:
Input: nums = [3,2,4], target = 6
Output: [1,2]

Example 3:
Input: nums = [3,3], target = 6
Output: [0,1]
```

````python
def find_target(array, target):
    if not len(array):
        raise ValueError
    idx = []
    if array[0] == target:
        idx.append(0)
    else:
        idx.append(-1)
    return idx
````

````python
%%ipytest -vv

def test_find_target_empty_input():
    with pytest.raises(ValueError):
        assert find_target([], 0)

def test_find_target_1_element_found():
    assert find_target([1], 1) == [0]

def test_find_target_1_element_not_found():
    assert find_target([1], 2) == [-1]
````

## Type hints for collections


````python
from typing import Any, Tuple, List, NamedTuple
````

````python
t1: Tuple[str, Any] = (1, 2)
s1: List[str] = ['a', 'b', 2]
````

````python
class Point(NamedTuple):
    x: float | int
    y: float | int

def distance(p1: Point, p2: Point) -> float:
    dx = p1.x - p2.x
    dy = p1.y - p2.y
    return (dx*dx + dy*dy) ** 0.5

a = Point(0,0)
b = Point(3,4)
assert distance(a, b) == 5
````

````python
from typing import NewType, TypeAlias, Optional, Union
````

````python
# Type alias

# version 1
# Number = Union[int, float]
# Array = List[Number]

#version 2
Number: TypeAlias = int | float
Array : TypeAlias = List[Number]

def mean_square_error(x: Array) -> float:
    return sum([v*v for v in x]) / len(x)

assert pytest.approx(mean_square_error([0.1, 0.2, 0.3]), rel=0.001) == 0.0467
````
